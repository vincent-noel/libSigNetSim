```python
%matplotlib notebook

from libsignetsim import SbmlDocument, EquilibriumPointCurve
from matplotlib import pyplot as plt
```

```python
doc = SbmlDocument()
doc.readSbmlFromFile("sbml_files/tyson2.xml")
```

```python
parameter = doc.model.listOfVariables.getBySbmlId('k4prime')
```

```python
curve = EquilibriumPointCurve(doc.model)
curve.setParameter(parameter)
curve.setRange(0, 5000)
curve.setDs(0.1)
curve.setMaxSteps(1000)
curve.build()
curve.run()
```

```python
variable_1 = doc.model.listOfVariables.getBySbmlId('cyclin_cdk_p')
variable_2 = doc.model.listOfVariables.getBySbmlId('cyclin_cdk')
```

```python
figure = plt.figure(figsize=(12, 6))
plot = figure.add_subplot(211)
plot.set_title(variable_1.getName())
curve.plotCurve(plot, variable_1)
if curve.hasHopfBifurcations():
    curve.plotLimitCycles(plot, variable_1)

plot = figure.add_subplot(212)
plot.set_title(variable_2.getName())
curve.plotCurve(plot, variable_2)
if curve.hasHopfBifurcations():
    curve.plotLimitCycles(plot, variable_2)
```
